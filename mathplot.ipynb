```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# making the first line diagram

```python
iris =  sns.load_dataset('iris')
mean_sepal = iris.groupby('species')['sepal_length'].mean()
plt.figure()
plt.plot(mean_sepal.index,mean_sepal.values,marker='o')
plt.title("Mean sepal length")
plt.xlabel("species")
plt.ylabel("Sepal_length")
plt
```

# using Seaborn 

```python
plt.figure
sns.lineplot(x='species',y='sepal_length',data=iris,estimator='mean')
plt.title("Mean sepeal leanght")
```

#### pie chart

```python
plt.figure()
plt.pie(mean_sepal.values,labels=mean_sepal.index,autopct='%1.1f%%')
plt.title("Mean sepal length distribution")
plt.show()
```

#### Seaborn(Histogram)

```python
plt.figure()
sns.histplot(iris['sepal_length'],bins=10)
plt.title("Distribution of sepal length")
plt.xlabel("sepal length")
plt.ylabel("Frequency")
plt.show()
```
