# Matplotlib using

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Math": [85, 90, 78, 92, 88],
    "Science": [80, 85, 82, 90, 87],
    "English": [78, 82, 80, 88, 85]
}
df = pd.DataFrame(data,index = ["Alice", "Bob", "Charlie", "David", "Eve"])
```

```python
## Ploting the data
```

```python
df.plot(kind="bar")
plt.title("Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend(title="Subjects")
plt.show()
```
