```python
df_cleaned = k.dropna() #drops the rows with missing values
df_cleaned.head()
```

```python
df_filled = k.fillna(0) #fills the missing values with 0
df_filled.head()
```

```python
df_group = k.groupby("day")["total_bill"].sum() #groups the data by day and calculates the mean of total_bill for each day
print(df_group.head())
```

```python
sorted_k = k.sort_values(by="total_bill", ascending=False) #sorts the dataframe by total_bill in descending order
print(sorted_k.head())
```

```python
k['total_bill_in hunders'] = k["total_bill"].apply(lambda x: x/100) #creates a new column total_bill_in hunders by applying a lambda function to total_bill column
print(k.head())
```

```python
k_merged = pd.merge(k, df_group, on="day", how="left") #merges the original dataframe with the grouped dataframe on the day column using left join
print(k_merged.head())
```
