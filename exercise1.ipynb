<br>Q Find the average electricity uses per hostel,the maximum usage per hostel and identufy which hostel has the highrst average uses

```python
average = np.mean(arr14,axis=1)
print("The average bill per hostel is :",average)
print("The highest bill per hostel is:",np.max(average,axis=0))
```

<b>Q2 Calculate the total electrucity bill for each hostel and find the highest bill among them.

```python
summ = np.sum(arr14,axis =0)
print("The total bill per day is:",summ)
print("The highest bill per day is:",np.max(summ,axis=0))
```
